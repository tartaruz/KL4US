## Import statements


```python
import pandas as pd
import os
filelist = os.listdir('dataset')
df_list = [pd.read_json('dataset/'+file, lines=True) for file in filelist]
df = pd.concat(df_list)
df
```

### Experimenting:

```python
df.stack().reset_index()
df.melt(id_vars=["eventId","userId","activeTime","documentId"], value_vars=["url"]).sort_values(by="userId")
```
